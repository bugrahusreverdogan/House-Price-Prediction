```python
temp = list(guzelyurt1[:])

for i in range(len(temp)):
    
    temp[i][8] = int(temp[i][8])
    temp[i][11] = int(temp[i][11])
    
    if '.' in temp[i][13]:
        idx = temp[i][13].index('.')
        temp[i][13] = temp[i][13][:idx]
    
    temp[i][0] = temp[i][0].replace('.', '')
    idx_1 = temp[i][0].find('TL')
    temp[i][0] = int(temp[i][0][:idx_1])
    
    if '-' in temp[i][7]:
        idx_2 = temp[i][7].find('-')
        temp[i][7] = int(temp[i][7][:idx_2])   
    
    elif ' ' in temp[i][7]:
        idx_2 = temp[i][7].find(' ')
        temp[i][7] = int(temp[i][7][:idx_2])    
    
    else:
        temp[i][7] = int(temp[i][7])
    
    idx = temp[i][4].index(' ') + 1
    idx2 = temp[i][4].index(' ', idx, len(temp[i][4])-1)
    temp[i][4] = temp[i][4][idx:idx2]
    
    idx_3 = temp[i][5].find(' ')
    temp[i][5] = int(temp[i][5][:idx_3])
    
    idx_3 = temp[i][6].find(' ')
    temp[i][6] = int(temp[i][6][:idx_3])
    
    if '+' in temp[i][12]:
        idx_4 = temp[i][12].find('+')
        temp[i][12] = int(temp[i][12][:idx_4]) + 1
    
    idx = temp[i][2].find(' ')
    temp[i][2] = float(temp[i][2][:idx])
    
    idx = temp[i][3].find(' ')
    temp[i][3] = float(temp[i][3][:idx])

guzelyurt1_modified = list(temp[:])
```
