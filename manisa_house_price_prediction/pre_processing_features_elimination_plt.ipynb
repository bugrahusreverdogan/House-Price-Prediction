### OUTLIER DETECTION WITH 3 SIGMA RULE

```python
data.columns
```

```python
for column in data.columns[:]:
    for nh in data["Neighborhood"].unique():
        selected_nh = data[data["Neighborhood"] == nh]
        selected_column = selected_nh[column]
        std = selected_column.std()
        avg = selected_column.mean()
        three_sigma_plus = avg + (3 * std)
        three_sigma_minus = avg - (3 * std)
        outliers = selected_column[(selected_nh[column] > three_sigma_plus) | (selected_nh[column] < three_sigma_minus)].index
        print("Neighborhood:", nh, column, outliers)
```

```python
for column in data.columns[:]:
    for nh in data["Neighborhood"].unique():
        selected_nh = data[data["Neighborhood"] == nh]
        selected_column = selected_nh[column]
        std = selected_column.std()
        avg = selected_column.mean()
        three_sigma_plus = avg + (3 * std)
        three_sigma_minus = avg - (3 * std)
        outliers = selected_column[(selected_nh[column] > three_sigma_plus) | (selected_nh[column] < three_sigma_minus)].index
        data.drop(index=outliers, inplace=True)
        print(nh, column, outliers)
```
